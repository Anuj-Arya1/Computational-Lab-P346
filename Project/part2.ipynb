```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
import os
import scipy.integrate as spi
from project_lib import *
```

```python
def rossler_attractor_equ(t, state, a, b, c):
    x, y, z = state
    dxdt = -y - z
    dydt = x + a * y
    dzdt = b + z * (x - c)
    return np.array([dxdt, dydt, dzdt])

for b in range(1,3):
    for a in range(1,4):
        sol = spi.solve_ivp(lambda t, y: rossler_attractor_equ(t, y,a*0.1,b*0.1,5.7  ),t_span=(0.0, 100.0),y0=[1.0, 1.0, 1.0],max_step=0.01)
        t_temp,x_temp,y_temp,z_temp = rossler_sol_lists(sol)
        plot_3d(x_temp, y_temp, z_temp,(a*0.1,b*0.1,5.7)  , 'C'+str(b-1))
        plot_time_series(t_temp, x_temp, y_temp, z_temp,(a*0.1,b*0.1,5.7))
        plot_spectrum(x_temp,(a*0.1,b*0.1,5.7),t_temp)









```
